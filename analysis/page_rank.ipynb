```python
import networkx as nx
import numpy as np
import pandas
import pylab
from pg import DB
db = DB()
```

```python
ms = db.query("SELECT shuffled_subgraph FROM motifss where shuffled_proteins_length = 36").getresult()[0][0]
m = db.query("SELECT subgraph FROM motifss where proteins_length = 36").getresult()[0][0]
distanceMatrix_df = pandas.read_pickle(open( "files/intermediates/distanceMatrix.pkl", "rb" ))
```

```python
norm = distanceMatrix_df.mean().mean()
pandas.DataFrame(m)
# norm
```

```python
for i, row in enumerate(m):
    for j, val in enumerate(row):
        if(val <= norm):
            print(i,j)
            m[i][j] = 0
        else:
            m[i][j] = 1
        if(i == j):
            m[i][j] = 0
```

```python
pandas.DataFrame(m)
```

```python
G = nx.from_numpy_matrix(np.array(m)) 

```

```python
ranks = nx.pagerank(G)
```
