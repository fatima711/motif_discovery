```python
import pandas
import pickle
import itertools
import re
```

```python
protein_to_sequence = pickle.load( open( "files/intermediates/proteins_to_sequences.pkl", "rb" ) )
protein_to_shuffled_sequence = pickle.load( open( "files/intermediates/proteins_to_shuffled_sequences.pkl", "rb" ) )
# protein_to_sequence #uncomment to inspect
```

The cell below generates all motifs. Each motif is genetated as an array. It will later need to be joined into strings

```python
alphabet = ["a","c", "g", "t"]
n = 8
allMotifs = list(itertools.product(alphabet,repeat=n)) #generates allmotifs of length n
# allMotifs #uncomment to inspect
len(allMotifs)
```

The cell below will associate motifs to the proteins

```python
%%time

motifs_to_proteins = {}
for motif in allMotifs:
    motif = "".join(motif) #turning motif array into string
    motifs_to_proteins[motif] = []
    for protein in protein_to_sequence:
#         regex = re.compile(motif)
        if(motif in protein_to_sequence[protein]):
            motifs_to_proteins[motif].append(protein)
```

Do the same for the shuffled dataset

```python
%%time

motifs_to_proteins_shuffled = {}
for motif in allMotifs:
    motif = "".join(motif) #turning motif array into string
    motifs_to_proteins_shuffled[motif] = []
    for protein in protein_to_shuffled_sequence:
#         regex = re.compile(motif)
        if(motif in protein_to_shuffled_sequence[protein]):
            motifs_to_proteins_shuffled[motif].append(protein)
```

```python
# print(len(motifs_to_proteins))
print(len(motifs_to_proteins_shuffled))
```

The cell bellow will save our protein to sequence in a python fiendly file

```python
fileName = "files/intermediates/motifs_to_proteins_7_to_" + str(n) + ".pkl"
pickle.dump(motifs_to_proteins, open( fileName, "wb" ))
```

```python
fileName = "files/intermediates/motifs_to_proteins_shuffled_4_to_" + str(n) + ".pkl"
pickle.dump(motifs_to_proteins_shuffled, open( fileName, "wb" ))
```

The cell bellow will just load the file, we just saves just to inspect and make sure it was correctly saved

```python
motifs_to_proteins = pickle.load( open( fileName, "rb" ) )
first2pairs = {k: motifs_to_proteins[k] for k in list(motifs_to_proteins)[:20]}
first2pairs
```
