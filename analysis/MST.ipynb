```python
import networkx as nx
import numpy as np
import pandas
import pylab
from pg import DB
db = DB()
```

```python
ms = db.query("SELECT shuffled_subgraph FROM motifss where shuffled_proteins_length = 36").getresult()[0][0]
m = db.query("SELECT subgraph FROM motifss where proteins_length = 36").getresult()[0][0]
db.query("SELECT shuffled_proteins_length FROM motifss order by shuffled_proteins_length asc").getresult()[0]

```

```python
for i, row in enumerate(m):
    for j, val in enumerate(row):
#         print(i,j)
        if(val == 1):
            m[i][j] = 0
#             print(m[i][j])
        else:
            m[i][j] = round(val, 2)
```

```python
G = nx.from_numpy_matrix(np.array(m)) 
Gs = nx.from_numpy_matrix(np.array(ms)) 


Ginv = nx.from_numpy_matrix(-np.array(m)) 

Gmin = nx.minimum_spanning_tree(G, weight = "weight")
Gmax  = nx.minimum_spanning_tree(Ginv, weight = "weight")

edge_labels=dict([((u,v,),d['weight'])
             for u,v,d in G.edges(data=True)])
Gmax
```
